What is Scatter Plot?


 Scatter plot is type of a plot that displays relationship between two variables.It shows how one variable is affected by another variables and helps identity trends, pattern,and correlation between the two variables.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('random variables')
plt.show()
```

what is subplot?

The subplot() function is used to plot multiple plots on the same figure in a grid-like pattern. This function allows us to easily create a set of subplots that are organized in rows and columns. It takes three arguments: the number of rows, the number of columns, and the index of the subplot we want to create.

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
```

```python
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y)
plt.show()
```

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
```

```python
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y)
plt.show()
```

What is Bar plot?Why it is used?

A bar plot is a type of plot that displays data as rectangular bars. Each bar represents a category, and the height (or length in case of horizontal bar plot) of the bar is proportional to the value of that category.

Bar plots are useful for comparing data across different categories, and are often used to visualize discrete data, such as counts or frequencies.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.xlabel('company')
plt.ylabel('profit')
plt.title("Company Profit")
plt.show()
```

```python
plt.barh(company,profit)
plt.xlabel('company')
plt.ylabel('profit')
plt.title('Company Profit')
plt.show()
```

What is Box Plot()?why it is used?

 box plot is a type of graph that displays the distribution of a dataset through their quartiles. The box in the plot represents the median value of the dataset, while the top and bottom edges of the box represent the upper and lower quartiles of the data respectively. Additionally, it displays outliers and extreme values, which are points that fall more than 1.5 times the interquartile (IQR) range away from the upper or lower quartile.

Box plots are helpful to quickly analyze the distribution of a single variable, identify outliers and compare multiple datasets side by side in a clear and concise manner.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
fig=plt.figure()
ax=fig.add_subplot()
box_plot_data = [box1, box2]
ax.boxplot(box_plot_data)
ax.set_title('Box Plot Example')
ax.set_xlabel('Value')
ax.set_ylabel('Box')
plt.show()
```
